Let us start by importing some packages which we will need below:

```python
import pandas as pd
import numpy as np
from scipy import optimize
from scipy.stats import norm
import datetime
import matplotlib.pyplot as plt
```

In this notebook, we calibrate an implied volatility surface free of static arbitrage opportunities to end--of--day bid and ask prices of options on the S&amp;P 500 index, traded on the CBOE on 10 January 2022. These options are European. Note that the level of the index on that day was 4670.29:

```python
SP500_20220110 = 4670.29
```

<h1>1. Loading the data.</h1>
Load the file containing the option price data into a <code>pandas</code> dataframe:

```python
df = pd.read_excel("spx_quotedata20220110_all.xlsx")
```

Inspect the first five rows of the dataframe:

```python
df.head()
```

Add columns "Call mid" and "Put mid" to this dataframe, containing the midpoint between the respective bid and ask prices:

```python
df["Call mid"]=(df["Bid"]+df["Ask"])/2
df["Put mid"]=(df["Bid.1"]+df["Ask.1"])/2
df.head()
```

Add a column "TTM" to this dataframe, containing the time to expiration of the options expressed as a fraction of a year (using the actual/365 daycount convention). Note the function `convdate()` defined to convert the string representation of the expiration date in the dataframe into a `datetime.date` object:

```python
monthdict = { "Jan":1,"Feb":2,"Mar":3,"Apr":4,"May":5,"Jun":6,"Jul":7,"Aug":8,"Sep":9,"Oct":10,"Nov":11,"Dec":12}
def convdate(s):
    return datetime.date(int(s.split()[3]),monthdict[s.split()[1]],int(s.split()[2]))

df["Expiration Date"] = df["Expiration Date"].apply(convdate)
today = datetime.date(2022, 1, 10)
df["TTM"] = df["Expiration Date"].apply(lambda d: (d-today).days/365)
df.tail()
```

<h1>2. Discount factors.</h1>
Assuming deterministic interest rates and dividends, for option pricing we need a term structure of zero coupon bond prices
\begin{equation}\label{bond}
  B(0,T)=\exp\left\{-\int_0^Tr(s)ds\right\}
\end{equation}
and "dividend discount factors"
\begin{equation}\label{div}
  D(0,T)=\exp\left\{-\int_0^Tq(s)ds\right\}
\end{equation}
Put/call parity asserts that for puts $P(K,T)$ and calls $C(K,T)$ of strike $K$ and maturity $T$ on the underlying $S$, it holds that
\begin{equation}\label{parity}
  C(K,T)-P(K,T)=D(0,T)S(0)-B(0,T)K
\end{equation}
Using the "Call mid" and "Put mid" prices, for each maturity in the data set, we can determine $B(0,T)$ and $D(0,T)$ such that the sum of squared deviations from put/call parity is minimised.

First, extract the times to maturity (greater than zero) for which we have option prices:

```python
TTM = df[df["TTM"]>0].TTM.unique()
```

Define a function calculating the sum of squared deviations from put/call parity:

```python
def f(x,C,P,K,S):
    D,B = x
    return np.sum((C-P-D*S+B*K)**2)
```

Iterate over all maturities $T$ to find the best-fitting $D(0,T)$ and $B(0,T)$:

```python
N      = len(TTM)
D      = np.empty(N)         # implied dividend discount factors
B      = np.empty(N)         # implied discount factors
i = 0
for T in TTM:
    C = df[df["TTM"]==T]["Call mid"].to_numpy()
    P = df[df["TTM"]==T]["Put mid"].to_numpy()
    K = df[df["TTM"]==T]["Strike"].to_numpy()
    args = C,P,K,SP500_20220110
    D[i],B[i] = optimize.minimize(f, [0.9,0.9], method='SLSQP',args=args).x
    i = i+1
```

<h1>3. Fitting a parametric implied volatility surface.</h1>
A parameterisation of the implied volatility surface that has been shown to work well with index options is (see also Gatheral and Jacquier (2014) and Gatheral (2013)):
    \begin{equation}\label{gatheral}
      w(k,T)=\frac{\theta_T}2\left(1+\rho\varphi(\theta_T)k+\sqrt{(\varphi(\theta_T)k+\rho)^2+(1-\rho)^2}\right)
    \end{equation}
    where $w(k,T)$ denotes the option <I>total implied variance</I>, $k$ is the log forward moneyness, $\theta_T$ is the forward at--the--money (i.e., $k=0$) total Black/Scholes implied variance for maturity $T$, $\varphi(\theta)$ is
    \begin{equation}\label{varphi}
      \varphi(\theta)=\frac{\eta}{\theta^{\gamma_1}(1+\beta_1\theta)^{\gamma_2}(1+\beta_2\theta)^{1-\gamma_1-\gamma_2}}
    \end{equation}
$\gamma_1 = 0.238$, $\gamma_2 = 0.253$, $\beta_1 = e^{5.18}$, $\beta_2 = e^{-3}$, $\eta=2.016048e^{\varepsilon}$ and $\varepsilon\in(-1,1)$ and $\rho$ are parameters.

Considering only options in the data set with a time to maturity of at least seven days, we want to determine the best-fitting $\rho$ and $\eta$ for this data set. "Best-fitting" in this context means that this fit minimises the sum of relative squared deviations of $w(k,T)$ outside the bounds of the total implied variance of the bid and ask option prices.

To this end, let's create a Python class <code>SSVI</code>, where the class constructor <code>__init__(self,df,S,minTTM)</code> takes as arguments the dataframe <code>df</code> containing the option price data, the current level of the underlying index <code>S</code> and the minimum time to maturity <code>minTTM</code>. Using these inputs, the class constructor calibrates an instance of <code>SSVI</code> to the market data. 

For this, we will need the functions to calculate the Black/Scholes implied volatility of an option in the case of deterministic dividends. Adapted from the code discussed in an earlier notebook, this is:

```python
def BlackScholes(S,K,sgm,r,q,T,callput):
    d1 = (np.log(S/K)+((r-q)+0.5*sgm*sgm)*T)/(sgm*np.sqrt(T))
    d2 = d1 - sgm*np.sqrt(T)
    return callput*(S*np.exp(-q*T)*norm.cdf(callput*d1)-K*np.exp(-r*T)*norm.cdf(callput*d2))

def BlackScholesSGM(sgm,S,K,r,q,T,callput,price):
    return price-BlackScholes(S,K,sgm,r,q,T,callput)

def BlackScholesImpVol(S,K,r,q,T,callput,price):
	sol = optimize.root_scalar(BlackScholesSGM, bracket=[0.0001, 1.0], args=(S,K,r,q,T,callput,price))
	return sol.root
```

The actual calibration is carried out by minimising an objective function, i.e. by calling `optimize.minimize()` on `self.objfunc`, where the member function `objfunc()` returns the squared relative deviation of the model $w(k,T)$ outside the bounds determined by the bid and ask prices. Denoting by $\sigma_{\text{bid}}(k,T)$ and $\sigma_{\text{ask}}(k,T)$ the Black/Scholes volatilities implied by the bid and ask option prices, respectively, the objective function is
$$
\sum_{i,j}\frac{\max(0,w(k_j,T_i)-\sigma^2_{\text{ask}}(k_j,T_i)T_i)^2+\min(0,w(k_j,T_i)-\sigma^2_{\text{bid}}(k_j,T_i)T_i)^2}{w(k_j,T_i)^2}
$$
where the sum is taken over all available log forward moneyness and maturities in the data set. The optimiser is tasked to look for the best-fitting $\rho$ and $\varepsilon$, where $\varepsilon$ is truncated to ensure it stays within $[-1,1]$.

Specifying the objective function of the calibration in this way means that quite a bit of preliminary work has to be carried out: Iterating over all maturities $T_i$, we use pu/call parity to determine `D[i]` and `B[i]`, from which we can calculate the interest rate `r[i]` and the dividend yield `q[i]`, as well as the forward price `F[i]`. For `theta[i]` (i.e., $\theta_{T_i}$), we use the total variance resulting from the average of the Black/Scholes implied volatilities of the call and the put with strike closest to the forward price. The total variances $\sigma^2_{\text{bid}}(k_j,T_i)T_i$ and $\sigma^2_{\text{ask}}(k_j,T_i)T_i$ implied by bid and ask prices are collected in the variables `TIVb` and `TIVa`, as the minimum of the call and put bid implied total variances and the maximum of the call and put ask implied total variances, respectively. `TIVb` and `TIVa` are lists (across all maturities) of NumPy arrays (across all strikes for a given maturity). 

Sometimes (especially away from the money) some market option prices are such that no valid Black/Scholes implied volatility can be found. This results in a `ValueError` exception, which is caught by the `try/except` block. In such cases, the data point is flaggeg by setting `self.k[i][j] = np.inf` and then removed from the calibration data set.

```python
class SSVI:
    def __init__(self,df,S,minTTM):
        self.S      = S
        self.minTTM = minTTM
        self.gamma1 = 0.238
        self.gamma2 = 0.253
        self.beta1  = np.exp(5.18)
        self.beta2  = np.exp(-3.0)
        self.TTM    = df[df["TTM"]>=minTTM].TTM.unique() # NumPy array
        self.N      = len(self.TTM)
        self.D      = np.empty(self.N)         # implied dividend discount factors
        self.B      = np.empty(self.N)         # implied discount factors
        self.q      = np.empty(self.N)         # implied dividend rates
        self.r      = np.empty(self.N)         # implied interest rates
        self.F      = np.empty(self.N)         # forward prices
        self.theta  = np.empty(self.N)         # at-the-money total implied variance
        self.ATMvol = np.empty(self.N)         # at-the-money volatility
        self.k      = []                       # log moneyness as list (in maturity) of arrays (in strikes)
        self.TIVCb  = []                       # total implied variance of call bid prices
        self.TIVCa  = []                       # total implied variance of call ask prices
        self.TIVPb  = []                       # total implied variance of put bid prices
        self.TIVPa  = []                       # total implied variance of put ask prices
        self.TIVb   = []                       # total implied variance of bid prices
        self.TIVa   = []                       # total implied variance of ask prices
        i = 0
        for T in self.TTM: # prepare data for calibration objective function
            C = df[df["TTM"]==T]["Call mid"].to_numpy()
            P = df[df["TTM"]==T]["Put mid"].to_numpy()
            K = df[df["TTM"]==T]["Strike"].to_numpy()
            # dividends and interest rates
            args = C,P,K
            self.D[i],self.B[i] = optimize.minimize(self.f, [0.9,0.9], method='SLSQP',args=args).x
            self.r[i] = -np.log(self.B[i])/T
            self.q[i] = -np.log(self.D[i])/T
            self.F[i] = self.S*self.D[i]/self.B[i]
            # at-the-money volatility
            idx = (np.abs(K - self.F[i])).argmin() # index of strike closest to ATM fwd
            v1 = BlackScholesImpVol(self.S,K[idx],self.r[i],self.q[i],T,1,C[idx])
            v2 = BlackScholesImpVol(self.S,K[idx],self.r[i],self.q[i],T,-1,P[idx])
            self.ATMvol[i] = (v1+v2)/2
            self.theta[i] = self.ATMvol[i]**2 * T
            # log forward moneyness
            self.k.append(np.log(K/self.F[i])) 
            # now calculate total implied variances from option prices
            Cb = df[df["TTM"]==T]["Bid"].to_numpy()
            Ca = df[df["TTM"]==T]["Ask"].to_numpy()
            Pb = df[df["TTM"]==T]["Bid.1"].to_numpy()
            Pa = df[df["TTM"]==T]["Ask.1"].to_numpy()
            CbTIV = np.empty(len(K))
            CaTIV = np.empty(len(K))
            PbTIV = np.empty(len(K))
            PaTIV = np.empty(len(K))
            bTIV  = np.empty(len(K))
            aTIV  = np.empty(len(K))
            for j in range(0,len(K)):
                if (np.abs(self.k[i][j])/self.theta[i]>3.5):
                    self.k[i][j] = np.inf
                else:             
                    try:                    
                        CbTIV[j] = BlackScholesImpVol(self.S,K[j],self.r[i],self.q[i],T,1,Cb[j])**2 * T                
                        CaTIV[j] = BlackScholesImpVol(self.S,K[j],self.r[i],self.q[i],T,1,Ca[j])**2 * T                
                        PbTIV[j] = BlackScholesImpVol(self.S,K[j],self.r[i],self.q[i],T,-1,Pb[j])**2 * T                
                        PaTIV[j] = BlackScholesImpVol(self.S,K[j],self.r[i],self.q[i],T,-1,Pa[j])**2 * T             
                        bTIV[j]  = np.minimum(CbTIV[j],PbTIV[j])                
                        aTIV[j]  = np.maximum(CaTIV[j],PaTIV[j])
                    except ValueError:
                        print("Skipping: ",self.S,K[j],self.r[i],self.q[i],T,1,Cb[j])
                        self.k[i][j] = np.inf
            self.TIVCb.append(CbTIV[self.k[i]!=np.inf])
            self.TIVCa.append(CaTIV[self.k[i]!=np.inf])
            self.TIVPb.append(PbTIV[self.k[i]!=np.inf])
            self.TIVPa.append(PaTIV[self.k[i]!=np.inf])
            self.TIVb.append(bTIV[self.k[i]!=np.inf])
            self.TIVa.append(aTIV[self.k[i]!=np.inf])
            self.k[i] = self.k[i][self.k[i]!=np.inf]
            i += 1
        epsilon,self.rho = optimize.minimize(self.objfunc, [0.5,-0.7], method='SLSQP').x
        self.set_eta(epsilon)
    def f(self,x,C,P,K):    # squared deviations from put/call parity
        D,B = x
        return np.sum((C-P-D*self.S+B*K)**2)
    def set_eta(self,epsilon):
        self.eta = 2.016048*np.exp(np.maximum(-1.0,np.minimum(1.0,epsilon)))
    def phi(self,theta):
        return self.eta/(theta**self.gamma1*(1+self.beta1*theta)**self.gamma2*(1+self.beta2*theta)**(1.0-self.gamma1-self.gamma2))
    def w(self,i,k):
        phi_k = self.phi(self.theta[i])*k
        return 0.5*self.theta[i]*(1+self.rho*phi_k+np.sqrt((phi_k+self.rho)**2+(1.0-self.rho**2)))
    def objfunc(self,x):    # calibration objective function
        self.set_eta(x[0])
        self.rho = x[1]
        phi = self.eta/(self.theta**self.gamma1*(1+self.beta1*self.theta)**self.gamma2*(1+self.beta2*self.theta)**(1.0-self.gamma1-self.gamma2))
        phi_k = phi*self.k
        w = 0.5*self.theta*(1+self.rho*phi_k+np.array(list(map(lambda i:np.sqrt(i),(phi_k+self.rho)**2+(1.0-self.rho**2))),dtype=object)) 
        err = np.sum(np.array(list(map(lambda i:np.sum(i),np.array(list(map(lambda x:np.maximum(0.0,x), w-self.TIVa)),dtype=object)**2/w**2)),dtype=object))
        err += np.sum(np.array(list(map(lambda i:np.sum(i),np.array(list(map(lambda x:np.minimum(0.0,x), w-self.TIVb)),dtype=object)**2/w**2)),dtype=object))
        return err
    def smile(self,i):      # return the model implied volatility smile for the i-th maturity
        phi_k = self.phi(self.theta[i])*self.k[i]
        return np.sqrt(0.5*self.theta[i]*(1+self.rho*phi_k+np.array(list(map(lambda i:np.sqrt(i),(phi_k+self.rho)**2+(1.0-self.rho**2)))))/self.TTM[i])
```

Now we can construct an instance `ssvi` of the `SSVI` class from the input market data (note the data points which are skipped because there is no valid Black/Scholes volatility implied by those market option prices):

```python
ssvi=SSVI(df,SP500_20220110,7.0/365)
```

The following code checks the quality of the calibration for each maturity $T_i$ by printing "OK" if the model total variance $w(k,T_i)$ lies between the bid and ask market-implied total variances. Otherwise, it prints bid, model and ask total variances for all available strikes for that maturity.

```python
n = len(ssvi.TIVb)
for i in range(0,n):
    phi = ssvi.eta/(ssvi.theta[i]**ssvi.gamma1*(1+ssvi.beta1*ssvi.theta[i])**ssvi.gamma2*(1+ssvi.beta2*ssvi.theta[i])**(1.0-ssvi.gamma1-ssvi.gamma2))
    phi_k = phi*ssvi.k[i]
    w = 0.5*ssvi.theta[i]*(1+ssvi.rho*phi_k+np.array(list(map(lambda i:np.sqrt(i),(phi_k+ssvi.rho)**2+(1.0-ssvi.rho**2)))))
    if (not(np.logical_and.reduce(ssvi.TIVb[i]<=w) and np.logical_and.reduce(w<=ssvi.TIVa[i]))):
        print(i,ssvi.TIVb[i],w,ssvi.TIVa[i])
    else:
        print(i,"OK")
```

Looking at this graphically, given $i$ the following code plots the bid, model and ask volatility smiles for maturity $T_i$ - feel free to change $i$ to inspect different maturities.

```python
i = 20
print("For maturity ",ssvi.TTM[i]*365," days:")
phi = ssvi.eta/(ssvi.theta[i]**ssvi.gamma1*(1+ssvi.beta1*ssvi.theta[i])**ssvi.gamma2*(1+ssvi.beta2*ssvi.theta[i])**(1.0-ssvi.gamma1-ssvi.gamma2))
phi_k = phi*ssvi.k[i]
w = 0.5*ssvi.theta[i]*(1+ssvi.rho*phi_k+np.array(list(map(lambda i:np.sqrt(i),(phi_k+ssvi.rho)**2+(1.0-ssvi.rho**2)))))
    
plt.plot(ssvi.k[i],np.sqrt(ssvi.TIVb[i]/ssvi.TTM[i]))
plt.plot(ssvi.k[i],np.sqrt(ssvi.TIVa[i]/ssvi.TTM[i]))
plt.plot(ssvi.k[i],np.sqrt(w/ssvi.TTM[i]))
```

```python
i = 11
print("For maturity ",ssvi.TTM[i]*365," days:")
smile = ssvi.smile(i)
    
plt.plot(ssvi.k[i],np.sqrt(ssvi.TIVb[i]/ssvi.TTM[i]))
plt.plot(ssvi.k[i],np.sqrt(ssvi.TIVa[i]/ssvi.TTM[i]))
plt.plot(ssvi.k[i],smile)
```

We see that even for maturities where the fit is not perfect, it is still quite good. In fact, the model smile crosses the bid and/or ask bounds primarily when we see a "sawtooth" pattern in the market bid and ask implied volatilities. This happens for times when the data set for a particular maturity includes weekly (SPXW) contracts as well as traditional (SPX) contracts, so this may have to do with market segmentation, i.e. a market imperfection not captured by the model:

# Question 1(a)
Below we have a Python function `AvgMC` which returns the Monte Carlo estimate for the Black/Scholes price of the option on the arithmetic average and the standard deviation of the simulation mean, where the function takes seven arguments (in this order): $S(0)$, $K$, $m$, $\sigma$, $r$, $T$ and the number of sampling iterations of the Monte Carlo algorithm $n$.

```python
import numpy as np

def AvgMC(S,K,m,sgm,r,T,n):
    dt = T/m # time step size
    acc = 0.0
    acc2 = 0.0
    for i in range(0,n): # loop over number of sampling iterations
        S_t = S
        avg = 0.0
        for j in range(0,m): # loop over time steps
            S_t = S_t*np.exp((r-0.5*sgm**2)*dt+sgm*np.sqrt(dt)*np.random.standard_normal())
            avg += S_t
        payoff=avg/m-K
        payoff=payoff*(payoff>0)
        acc = acc + payoff
        acc2 = acc2 + payoff*payoff
    # Monte Carlo estimate
    MC=np.exp(-r*T)*acc/n
    # Standard deviation of the simulation mean
    MCstd=np.exp(-r*T)*np.sqrt(acc2/n-acc*acc/n**2)/np.sqrt(n)
    return MC, MCstd
```

# Question 1(b)
Below we have a Python function `GeoAvgOption` which implements the closed-form solution for an option on the discretely observed geometric average, where the function takes six arguments (in this order): $S(0)$, $K$, $m$, $\sigma$, $r$ and $T$.

```python
from scipy.stats import norm

def tautilde(T,m,dt):
    return T-dt*(m-1)/2

def tau(T,m,dt):
    return T-dt*m*(m-1)*(4*m+1)/(6*m*m)

def GeoAvgOption(S,K,m,sgm,r,T):
    dsa = (np.log(S/K)+(r-0.5*sgm**2)*tautilde(T,m,T/m))/(sgm*np.sqrt(tau(T,m,T/m)))
    return S*np.exp(-r*(T-tautilde(T,m,T/m))-sgm**2*(tautilde(T,m,T/m)-tau(T,m,T/m))/2)*norm.cdf(dsa+sgm*np.sqrt(tau(T,m,T/m)))-K*np.exp(-r*T)*norm.cdf(dsa)
```

# Question 1(c)
Below we have a Python function `AvgMCCV` which returns the Monte Carlo estimate for the Black/Scholes price of the option on the arithmetic average and the standard deviation of the simulation mean, using the option on the geometric average as control variate. The function takes seven arguments (in this order): $S(0)$, $K$, $m$, $\sigma$, $r$, $T$ and the number of sampling iterations of the Monte Carlo algorithm $n$.

```python
import numpy as np

def AvgMCCV(S,K,m,sgm,r,T,n):
    dt = T/m # time step size
    acc = 0.0
    acc2 = 0.0
    for i in range(0,n): # loop over number of sampling iterations
        lnS_t = np.log(S)
        avg = 0.0
        geo = 0.0
        for j in range(0,m): # loop over time steps
            lnS_t += ((r-0.5*sgm**2)*dt+sgm*np.sqrt(dt)*np.random.standard_normal())
            geo += lnS_t # accumulating the geometric average in log terms
            S_t = np.exp(lnS_t)
            avg += S_t
        # payoff of option on the arithmetic average
        payoff=avg/m-K
        payoff=payoff*(payoff>0)
        # payoff of option on the geometric average
        payoff2=np.exp(geo/m)-K
        payoff2=payoff2*(payoff2>0)
        # payoff difference
        payoff -= payoff2
        acc = acc + payoff
        acc2 = acc2 + payoff*payoff
    # Monte Carlo estimate
    MC=np.exp(-r*T)*acc/n + GeoAvgOption(S,K,m,sgm,r,T)
    # Standard deviation of the simulation mean
    MCstd=np.exp(-r*T)*np.sqrt(acc2/n-acc*acc/n**2)/np.sqrt(n)
    return MC, MCstd
```

Run the Monte Carlo routines without and with the control variate:

```python
S=390
K=400
sgm=0.2
r=0.07
T=1
m=12
n=100000
np.random.seed(1234)
MC, MCstd=AvgMC(S,K,m,sgm,r,T,n)
print([n,MC,MCstd,MC-2*MCstd,MC+2*MCstd])
np.random.seed(1234)
MC, MCstd=AvgMCCV(S,K,m,sgm,r,T,n)
print([n,MC,MCstd,MC-2*MCstd,MC+2*MCstd])
```

The key difference is that the standard deviation of the simulation mean is much lower when using the control variate, resulting in a tighter confidence interval (the last two numbers in each line above), i.e., a higher accuracy of the Monte Carlo estimate.

# Question 2(a)
We want to implement the option pricing formula in the shifted mixture-of-lognormals model,
$$
\mathcal{O}=\omega P(0,T)\sum_{i=1}^N\lambda_i\Biggl(\mathcal{A}_0e^{\mu T}\Phi\left(\omega\frac{\ln\frac{\mathcal{A}_0}{\mathcal{K}}+(\mu+\frac12\eta_i^2)T}{\eta_i\sqrt{T}}\right)\\
-{\mathcal{K}} \Phi\left(\omega\frac{\ln\frac{\mathcal{A}_0}{\mathcal{K}}+(\mu-\frac12\eta_i^2)T}{\eta_i\sqrt{T}}\right)\Biggr)
$$
noting that
\begin{eqnarray*}
{\mathcal{K}} &=& K-A_0\alpha e^{\mu T}\\
{\mathcal{A}_0} &=& A_0(1-\alpha)
\end{eqnarray*}
This formula involves a sum over Black/Scholes formulas, so we use a copy of the Python implementation of this formula from the lectures:

```python
def BlackScholes(S,K,sgm,r,T,callput):
    d1 = (np.log(S/K)+(r+0.5*sgm*sgm)*T)/(sgm*np.sqrt(T))
    d2 = d1 - sgm*np.sqrt(T)
    return callput*(S*norm.cdf(callput*d1)-K*np.exp(-r*T)*norm.cdf(callput*d2))
```

This then allows us to implement the option pricing formula in the shifted mixture-of-lognormals model:

```python
def LognormalMixtureOption(A0,K,eta,lam,r,T,alpha,callput):
    result = 0.
    calA = A0*(1-alpha)
    calK = K-A0*alpha*np.exp(r*T)
    for i in range(0,len(eta)):
        result += lam[i]*BlackScholes(calA,calK,eta[i],r,T,callput)
    return result
```

# Question 2(b)
Loading the option data file `GOOG.xlsx`:

```python
import pandas as pd
GOOGopt=pd.read_excel('GOOG.xlsx',index_col=False,parse_dates=[3,6])
GOOGopt.head()
```

Extracting those call options (only) expiring on the 17th of June 2006 (only), for which the trading volume was greater than 100:

```python
df = GOOGopt.loc[GOOGopt['call_put'] == 'C'] 
df = df.loc[(df['option_expiration'] == '2006-06-17')] 
df = df.loc[df['volume'] > 100] 
df
```

To plot the implied volatility smile for these options, we can use a copy of the Python implementation for the implied volatility calculation from the lectures:

```python
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline

def BlackScholesSGM(sgm,S,K,r,T,callput,price):
    return price-BlackScholes(S,K,sgm,r,T,callput)

def BlackScholesImpVol(S,K,r,T,callput,price):
    sol = opt.root_scalar(BlackScholesSGM, bracket=[0.0001, 1], args=(S,K,r,T,callput,price))
    return sol.root
```

Plotting the smile as demonstrated in the lectures (the 'bid' and 'ask' implied volatilities are not strictly required in this question, but will be needed in 2(c)):

```python
# Extract initial stock price
S0 = df['stock_price_for_iv'][df.index[0]]
# Extract time to maturity
tau = (df['option_expiration'][df.index[0]]-df['date'][df.index[0]]).days/365
# Calculate continuously compounded interest rate
r=np.log(1+0.05/4)*4

# Calculate the implied volatilities
smile = pd.DataFrame(columns=['Strike','Implied vol','Implied vol bid','Implied vol ask'])
smile = smile.set_index('Strike')
for x in df.index:
    K = df['strike'][x]
    price = (df['ask'][x]+df['bid'][x])/2.
    smile.loc[K] = (BlackScholesImpVol(S0,K,r,tau,1,price),BlackScholesImpVol(S0,K,r,tau,1,df['bid'][x]),BlackScholesImpVol(S0,K,r,tau,1,df['ask'][x]))

# Create plot
plt.plot(smile['Implied vol'])
plt.plot(smile['Implied vol bid'])
plt.plot(smile['Implied vol ask'])
plt.show()
```

# Question 2(c)
Using SciPy `minimize()` with `method='Nelder-Mead'`, calibrate the parameters of a model with $N=2$. First, we set up an objective function to perform the calibration using the SciPy `minimize(...)` function. This objective function returns the sum of the squared relative violations of the bid/ask bounds given by the market, i.e.
$$
\sum_j \frac{\max(0,C_{\text{model}}-C_{\text{ask}})^2}{C_{\text{ask}}^2}+
\frac{\max(0,C_{\text{bid}}-C_{\text{model}})^2}{C_{\text{bid}}^2}
$$

```python
def objective_function(par,args):
    S,r,ttm,df,dim = args
    eta = par[:dim]
    lam = list(par[dim:dim*2-1])
    lam.append(1.-sum(lam))
    alpha=par[dim*2-1]
    err = 0.0
    for x in df.index:
        strike = df['strike'][x]
        price  = LognormalMixtureOption(S,strike,eta,lam,r,ttm,alpha,1)
        if (price>df['ask'][x]): 
            err += ((price-df['ask'][x])/df['ask'][x])**2
        if (price<df['bid'][x]): 
            err += ((price-df['bid'][x])/df['bid'][x])**2
    return err
```

Now we run the calibration:

```python
par = 0.3,0.2,0.5,0.1 # starting point for the calibration search
result=opt.minimize(objective_function,par,args=[S0,r,tau,df,2],method='Nelder-Mead',tol=0.01)
```

Plot the implied volatility smile given by the market bid, the market ask, and the implied volatility smile given by the model in a single graph:

```python
# extract the calibration results
dim=2
eta=result.x[:dim]
lam = list(result.x[dim:dim*2-1])
lam.append(1.-sum(lam))
alpha=result.x[-1]

# plot the implied volatilities
strikelist = smile.index
impvollist = []
for strike in strikelist:
    price  = LognormalMixtureOption(S0,strike,eta,lam,r,tau,alpha,1)
    impvol = BlackScholesImpVol(S0,strike,r,tau,1,price)
    impvollist.append(impvol)
plt.plot(strikelist,impvollist)
plt.plot(smile['Implied vol'])
plt.plot(smile['Implied vol bid'])
plt.plot(smile['Implied vol ask'])
plt.show()
```
